```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import os 

os.getcwd()
os.listdir(os.getcwd())
```

```python
path1 = "Pymaceuticals_data_clinicaltrial_data.csv"
path2 = "Pymaceuticals_data_mouse_drug_data.csv"

mouse_data = pd.read_csv(path1)
trial_data = pd.read_csv(path2)
```

```python
mouse_data.head(2)
```

```python
trial_data.head(2)
```

```python
combined_df = pd.merge(mouse_data, trial_data, how="left", on="Mouse ID")
combined_df
```

```python
combined_df.columns
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
dt = combined_df.iloc[:,[1,2,4]]
dt

dt_grouped = dt.groupby(['Drug', 'Timepoint'])
dt_grouped_mean = dt.groupby(['Drug', 'Timepoint']).mean()
dt_grouped_mean

# Convert to DataFrame
dt_df = pd.DataFrame(dt_grouped_mean)

# Preview DataFrame
dt_df.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
dt_grouped_se = dt.groupby(['Drug', 'Timepoint'])
dt_grouped_se = dt_grouped_se.sem()

# Convert to DataFrame
se_df = pd.DataFrame(dt_grouped_se)

# Preview DataFrame
se_df = se_df.reset_index()
se_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

munging = se_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
munging = pd.DataFrame(munging)
munging.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(x=munging.index,y=dt_df.loc["Capomulin"]["Tumor Volume (mm3)"],marker = 'o', yerr=munging.Capomulin);
plt.errorbar(x=munging.index,y=dt_df.loc["Infubinol"]["Tumor Volume (mm3)"],marker = '1', yerr=munging.Infubinol);
plt.errorbar(x=munging.index,y=dt_df.loc["Ketapril"]["Tumor Volume (mm3)"],marker = 's',yerr=munging.Ketapril);
plt.errorbar(x=munging.index,y=dt_df.loc["Placebo"]["Tumor Volume (mm3)"],marker = '*',yerr=munging.Placebo);

plt.title("Tumor Response to Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

plt.grid()
plt.show()

# Save the Figure
plt.savefig("../Pymaceuticals/Plot1.png")
plt.show()
```

```python
# Calculate the percent changes for each drug
Captt0 = dt_df.iloc[0]
Captt45 = dt_df.iloc[9]

Ceftt0 = dt_df.iloc[10]
Ceftt45 = dt_df.iloc[19]

Inftt0 = dt_df.iloc[20]
Inftt45 = dt_df.iloc[29]

Kettt0 = dt_df.iloc[30]
Kettt45 = dt_df.iloc[39]

Naftt0 = dt_df.iloc[40]
Naftt45 = dt_df.iloc[49]

Platt0 = dt_df.iloc[50]
Platt45 = dt_df.iloc[59]

Prott0 = dt_df.iloc[60]
Prott45 = dt_df.iloc[69]

Ramtt0 = dt_df.iloc[70]
Ramtt45 = dt_df.iloc[79]

Stett0 = dt_df.iloc[80]
Stett45 = dt_df.iloc[89]

Zontt0 = dt_df.iloc[90]
Zontt45 = dt_df.iloc[99]

CapPercent = ((Captt45-Captt0)/Captt0)*100
CefPercent = ((Ceftt45-Ceftt0)/Ceftt0)*100
InfPercent = ((Inftt45-Inftt0)/Inftt0)*100
KetPercent = ((Kettt45-Kettt0)/Kettt0)*100
NafPercent = ((Naftt45-Naftt0)/Naftt0)*100
PlaPercent = ((Platt45-Platt0)/Platt0)*100
ProPercent = ((Prott45-Prott0)/Prott0)*100
RamPercent = ((Ramtt45-Ramtt0)/Ramtt0)*100
StePercent = ((Stett45-Stett0)/Stett0)*100
ZonPercent = ((Zontt45-Zontt0)/Zontt0)*100

pc = [CapPercent, CefPercent, InfPercent, KetPercent, NafPercent, PlaPercent, ProPercent, RamPercent, StePercent, ZonPercent]
pc = pd.DataFrame(pc)
pc = pc.rename(columns={"Tumor Volume (mm3)":"Percent Change in Tumor Volume (mm3)"})
pc = pc.rename(index={0:"Capomulin", 1:"Ceftamin", 2:"Infubinol",3:"Ketapril", 4:"Naftisol",5:"Placebo", 6:"Propriva",7:"Ramicane",8:"Stelasyn",9:"Zoniferol"})
pc
```

```python
# Store all Relevant Percent Changes into a Tuple
pct = ([pc.iloc[0], pc.iloc[2], pc.iloc[3], pc.iloc[5]])
pct

# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
# Call functions to implement the function calls
fig, ax = plt.subplots()

ind = np.arange(len(pct)) 
width = 1
rectsPass = ax.bar(ind[0], pct[0], color='green')
rectsFail1 = ax.bar(ind[1], pct[1], color='red')
rectsFail2 = ax.bar(ind[2], pct[2], color='red')
rectsFail3 = ax.bar(ind[3], pct[3], color='red')

ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., 3,
               '%d%%' % int(height),
               ha='center', va='bottom', color="white")
def autolabelPass(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., -8,
               '-%d%% ' % int(height),
               ha='center', va='bottom', color="white")
    
autolabelPass(rectsPass)
autolabelFail(rectsFail1)
autolabelFail(rectsFail2)
autolabelFail(rectsFail3)
# Save the Figure

# Show the Figure
```
